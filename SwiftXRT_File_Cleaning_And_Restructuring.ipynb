```python
import re
import decimal
import pandas as pd
import warnings
import os
warnings.filterwarnings("ignore")
```

```python
head, filename = os.path.split("/content/flux_plain_2.dat")
```

```python
with open(filename) as file:
  data = file.read()
```

```python
data
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
filenew = open(filename,'w+')
```

```python
Cols = ['!Time','T_+ve','T_-ve','Flux','Fluxpos','Fluxneg']
```

```python
for x in Cols:
  filenew.write(x)
  filenew.write('\t')
filenew.write('\n')
```

```python
temp = ''
check = False
specals = [':','/','.','!','-','+']
c = 0
for x in data :
  if (x.isalpha() or x.isdigit() or x in specals):
    temp += x
  else :
    if temp == Cols[5]:
      check = True
    if(is_number(temp) and check):
      filenew.write(temp)
      filenew.write('\t')
      c += 1
    temp = ''
  if c == 6:
    filenew.write('\n')
    c = 0
```

```python
filenew.close()
```
